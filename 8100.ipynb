```julia
function limit_geo2(n=10^3)
    X = []
    for i = 1:13
        push!(X,i,i,i,i)
    end
    Z = []
    W = []
    k = 0
    while length(X) ≥ 4
        Y= rand(X,4)
        k += 1
        if Y[1] ≠ Y[2] ≠ Y[3] ≠ Y[4] ≠ Y[2] && Y[3] ≠ Y[1] ≠ Y[4] 
            if X |> length ==52
                append!(Z,Y)
                for y in Y
                    for i = 1:length(X)
                        if X[i] == y
                            deleteat!(X,i)
                            break
                        end
                    end
                end
                push!(W,Y)
            elseif all(x -> length(x ∩ Z) == 1 , Z)
                append!(Z,Y)
                for y in Y
                    for i = 1:length(X)
                        if X[i] == y
                            deleteat!(X,i)
                            break
                        end
                    end
                end
                push!(W,Y)
            end
        end
        if  k == n
            break
        end
    end
    k,X,W
end
```

```julia
A = [limit_geo2(10^2)[1] for _=1:100]
using Plots,Statistics
scatter(A,label="ratio 100 vals $(count(x -> x == 100 , A)/100)")
```
