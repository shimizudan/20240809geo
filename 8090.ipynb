```julia
function limit_geo()
    X = []
    for i = 1:13
        push!(X,i,i,i,i)
    end
    Z = []
    W = []
    k = 0
    while length(X) ≥ 4
        Y= rand(X,4)
        k += 1
        if Y[1] ≠ Y[2] ≠ Y[3] ≠ Y[4] ≠ Y[2] && Y[3] ≠ Y[1] ≠ Y[4] 
            if X |> length ==52
                append!(Z,Y)
                for y in Y
                    for i = 1:length(X)
                        if X[i] == y
                            deleteat!(X,i)
                            break
                        end
                    end
                end
            elseif length(Y ∩ Z) == 1
                append!(Z,Y)
                for y in Y
                    for i = 1:length(X)
                        if X[i] == y
                            deleteat!(X,i)
                            break
                        end
                    end
                end
            end
        end
        push!(W,Y => Z|> Set)
        if Z |> Set |>length ==13 || k == 10^4
            break
        end
    end
    W,k
end

```
